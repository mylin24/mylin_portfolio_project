### Which stores have the lowest and highest unemployment rates?  ###

```python
min_unemployment = df.groupby('Store')['Unemployment'].min()
max_unemployment = df.groupby('Store')['Unemployment'].max()
# Stores have the lowest and highest unemployment rate
store_with_min_employment= min_unemployment.idxmin(),  min_unemployment.min()
store_with_max_employment= max_unemployment.idxmax(), max_unemployment.max()
store_with_min_employment, store_with_max_employment
```

The result shows that the store with the lowest unemployment rate is Store 4 with 3.879%  and the store with the highest  unemployment rate is Store 12 with 14.313%.


```python
Unemployment_by_store = df.groupby('Store')['Unemployment'].mean().reset_index()
min_unemployment_store = Unemployment_by_store.loc[Unemployment_by_store['Unemployment'].idxmin()]
max_unemployment_store = Unemployment_by_store.loc[Unemployment_by_store['Unemployment'].idxmax()]
print(f"Store with the lowest unemployment rate is:{min_unemployment_store['Store']} with the rate: {min_unemployment_store['Unemployment']}%")
print(f"Store with the highest unemployment rate is:{max_unemployment_store['Store']} with the rate: {max_unemployment_store['Unemployment']}%")
# Vizualization
plt.figure(figsize=(12,6))
sns.barplot(x='Store', y='Unemployment', data=Unemployment_by_store, palette = 'viridis')
plt.axhline(min_unemployment_store['Unemployment'],color='blue', linestyle='--', label = f"Lowest Unemployment:{min_unemployment_store['Store']}")
plt.axhline(max_unemployment_store['Unemployment'],color ='red', linestyle='--', label = f"Highest Unemployment:{max_unemployment_store['Store']}") 
plt.title('Average Unemployment Rate by Store')
plt.xlabel('Store')
plt.ylabel('Average Unemployment Rate')
plt.xticks(rotation=45) # Xoay nhÃ£n trÃªn trá»¥c X gÃ³c 45
plt.legend()
plt.show()
```

### Is there any correlation between CPI and Weekly Sales? How is the correlation different when the Holiday Flag is 0 versus when the Holiday Flag is 1? ###

```python
Corr_non_holiday = df[df['Holiday_Flag']==0][['CPI','Weekly_Sales']].corr()
Corr_hodiday= df[df['Holiday_Flag']== 1][['CPI', 'Weekly_Sales']].corr()
print('Correlation between CPI and Weekly Sales when the Holiday Flag is 0',Corr_non_holiday)
print('Correlation between CPI and Weekly Sales when the Holiday Flag is 1',Corr_hodiday)
```

From the result above, we can see that the correlation between CPI and Weekly Sales when there are  Holidays and there are no  Holidays are both negative correlations with the similar values. This means that when CPI increses, Weekly sales tend to decrease and vice versa. When CPI increases, the average price rise, leading to inflation, meaning the cost of living is rising.
As prices rise due to inflation, consumers' purchasing power diminishes. They may find that their income doesn't stretch as far, leading them to cut back on non-essential purchases or switch to cheaper alternatives. This reduction in consumer spending can lead to a decrease in weekly sales.
